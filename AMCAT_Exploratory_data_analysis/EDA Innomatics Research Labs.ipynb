```python
numerical_columns = ['Salary', '10percentage', '12percentage']
for column in numerical_columns:
    plt.figure(figsize=(8, 5))
    sns.histplot(df[column], kde=True)
    plt.title(f'Histogram and PDF for {column}')
    plt.show()
```

```python
categorical_columns = ['Gender', 'Designation', 'JobCity']
for column in categorical_columns:
    plt.figure(figsize=(10, 6))
    sns.countplot(y=df[column], order=df[column].value_counts().index)
    plt.title(f'Countplot for {column}')
    plt.show()
```

```python
def numerical_univariate_analysis(numerical_data):
    for col_name in numerical_data:
        print("#" * 10, col_name, "#" * 10)
        print(numerical_data[col_name].agg(['min', 'max', 'mean', 'median', 'std', 'skew', 'kurt']))
        print()
```

```python
numerical_univariate_analysis(df[['ID','Salary']])
```

```python
def discrete_univariate_analysis(discrete_data):
    for col_name in discrete_data:
        print("#" * 10, col_name, "#" * 10)
        print(discrete_data[col_name].agg(['count', 'nunique', 'unique']))
        print('value counts: \n', discrete_data[col_name].value_counts(normalize=True))
        print()
```

```python
discrete_univariate_analysis(df[['Gender','JobCity']])
```

```python
#BIVARIRTE ANALYSIS
plt.figure(figsize=(8, 5))
sns.scatterplot(x='10percentage', y='Salary', data=df)
plt.title('Scatter plot between Salary and 10percentage')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x='Gender', y='Salary', data=df)
plt.title('Boxplot of Salary by Gender')
plt.show()
```

```python
cross_tab = pd.crosstab(df['Gender'], df['Designation'])
cross_tab.plot(kind='bar', stacked=True, figsize=(10, 7))
plt.title('Stacked Bar Plot between Gender and Designation')
plt.show()
```

```python
roles = ['Programming Analyst', 'Software Engineer', 'Hardware Engineer', 'Associate Engineer']
filtered_data = df[df['Designation'].isin(roles)]
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(y='Designation', x='Salary', data=filtered_data)
plt.title('Salary Distribution for Fresh Graduates in Specified Roles')
plt.show()
```
